```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("result.csv", sep=';')

t = df['t']
x = df['x']
y = df['y']
z = df['z']

plt.figure(figsize=(20,10))

plt.plot(t,x, label='Acelerao no eixo X', color='blue')
plt.plot(t,y, label='Acelerao no eixo Y', color='red')
plt.plot(t,z, label='Acelerao no eixo Z', color='orange')

plt.title('Grfico da acelerao em funo do tempo')
plt.xlabel('Tempo (s)')
plt.ylabel('Acelerao (g)')
plt.legend()

plt.tight_layout()
plt.show()
```
